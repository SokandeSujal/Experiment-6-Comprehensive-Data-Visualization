In this notebook, we will explore various aspects of data visualization using UCI ML datasets and integrate Matplotlib with other libraries. The content includes data loading and cleaning, basic and advanced visualizations, library integration, and best practices.

## **1. Data Visualization with UCI ML Datasets**

### **1.1. Data Loading and Cleaning**




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset from UCI ML repository
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
data = pd.read_csv(url, header=None, names=column_names)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Data exploration
print("\nSummary statistics:")
print(data.describe())

print("\nChecking for missing values:")
print(data.isnull().sum())

# Data cleaning (if needed)
# In this dataset, there are no missing values or obvious data type issues.
```

### **1.2. Basic Visualizations**

```python
# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(data['sepal_length'], marker='o', linestyle='-', color='b', label='Sepal Length')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.title('Line Plot of Sepal Length')
plt.legend()
plt.grid(True)
plt.savefig('Line_Plot.png')
plt.show()

# Bar Plot
species_count = data['species'].value_counts()
plt.figure(figsize=(8, 5))
species_count.plot(kind='bar', color='green')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Plot of Species Count')
plt.savefig('Bar_Plot.png')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data['sepal_length'], data['sepal_width'], c=data['species'].astype('category').cat.codes, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Dimensions')
plt.colorbar(label='Species')
plt.savefig('Scatter_Plot.png')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(data['petal_length'], bins=20, color='orange', edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.savefig('Histogram.png')
plt.show()

# Pie Chart
species_pie = data['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_pie, labels=species_pie.index, autopct='%1.1f%%', colors=['red', 'blue', 'green'])
plt.title('Pie Chart of Species Distribution')
plt.savefig('Pie_Chart.png')
plt.show()
```

## **1.3. Advanced Visualizations**

```python
import seaborn as sns

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=data)
plt.title('Box Plot of Petal Length by Species')
plt.savefig('Box_Plot.png')
plt.show()

# Heatmap
correlation_matrix = data.drop('species', axis=1).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Feature Correlations')
plt.savefig('Heatmap.png')
plt.show()

# Contour Plot
x = np.linspace(data['sepal_length'].min(), data['sepal_length'].max(), 100)
y = np.linspace(data['sepal_width'].min(), data['sepal_width'].max(), 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2))  # Example function for demonstration

plt.figure(figsize=(10, 6))
contour = plt.contourf(X, Y, Z, cmap='plasma')
plt.colorbar(contour)
plt.title('Contour Plot Example')
plt.savefig('Contour_Plot.png')
plt.show()

# 3D Plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['sepal_length'], data['sepal_width'], data['petal_length'], c=data['species'].astype('category').cat.codes, cmap='viridis')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
plt.title('3D Scatter Plot of Sepal and Petal Dimensions')
plt.savefig('3D_Plot.png')
plt.show()
```

# **2. Integration with Other Libraries**

## **2.1. Pandas Integration**

```python
# Simple Line Plot with Pandas
data[['sepal_length', 'sepal_width']].plot(kind='line', figsize=(10, 6))
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot with Pandas')
plt.grid(True)
plt.savefig('Pandas_Line_Plot.png')
plt.show()

# Histogram with Pandas
data['petal_width'].plot(kind='hist', bins=20, figsize=(10, 6), color='purple', edgecolor='black')
plt.xlabel('Petal Width (cm)')
plt.title('Histogram of Petal Width with Pandas')
plt.savefig('Pandas_Histogram.png')
plt.show()
```

## **2.2. Seaborn Integration**

```python
# Pair Plot with Seaborn
sns.pairplot(data, hue='species', palette='husl')
plt.title('Pair Plot of Iris Dataset')
plt.savefig('Pair_Plot.png')
plt.show()

# Customizing Seaborn Plot with Matplotlib
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='petal_length', data=data, palette='muted')
plt.title('Violin Plot of Petal Length by Species')
plt.grid(True)
plt.savefig('Violin_Plot.png')
plt.show()
```

# **3. Real-world Applications and Best Practices**

## **3.1. Visualization Best Practices**
### **Principles of Good Visualization**

1. **Clarity:** Ensure that the visualization clearly communicates the message.
2. **Accuracy:** Present data accurately without distortion.
3. **Relevance:** Use appropriate charts for the data type and context.
4. **Simplicity:** Avoid unnecessary decorations and focus on the data.
5. **Consistency:** Maintain consistency in styles, colors, and scales.

### **Examples of Poor and Good Visualizations**

### **Poor Visualization:** Overcomplicated with too many elements and unclear axes.
### **Good Visualization:** Clean and straightforward with clearly labeled axes and a focus on the data.

## **3.3. Projects**

```python
# Example project: Combining multiple plots for a comprehensive analysis
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line Plot
axs[0, 0].plot(data['sepal_length'], marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Sepal Length')

# Histogram
axs[0, 1].hist(data['petal_length'], bins=20, color='orange', edgecolor='black')
axs[0, 1].set_title('Petal Length Distribution')

# Box Plot
sns.boxplot(x='species', y='petal_width', data=data, ax=axs[1, 0])
axs[1, 0].set_title('Box Plot of Petal Width')

# Pie Chart
data['species'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['red', 'blue', 'green'], ax=axs[1, 1])
axs[1, 1].set_title('Species Distribution')

plt.tight_layout()
plt.savefig('Comprehensive_Project.png')
plt.show()
```

# **4. Advanced Topics (Optional)**

## **4.1. Animations**

```python
import matplotlib.animation as animation

# Example animated plot
fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x), 'r-')

def animate(i):
    line.set_ydata(np.sin(x + i / 10.0))  # Update data
    return line,

ani = animation.FuncAnimation(fig, animate, frames=100, interval=50, blit=True)
plt.title('Animated Sine Wave')
plt.show()
```

## **4.2. Custom Figures**

```python
# Example of a custom figure design
fig, ax = plt.subplots(figsize=(10, 6))

# Custom plot with complex design
ax.plot(data['sepal_length'], data['sepal_width'], 'o-', color='magenta', markersize=8, linewidth=2)
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_title('Customized Plot with Complex Design')
plt.grid(True)

plt.savefig('Custom_Figure.png')
plt.show()
```

## **4.3. Interactive Plots**

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Cursor

fig, ax = plt.subplots()
ax.plot(data['sepal_length'], data['sepal_width'], 'o-', color='blue')

cursor = Cursor(ax, useblit=True, color='red', linewidth=1)
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_title('Plot with Interactive Cursor')
plt.show()
```
